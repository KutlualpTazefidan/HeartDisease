```python
# Importing Libraries
import pandas as pd
```

```python
feature_names = ['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
```

```python
df = pd.read_csv('./HeartDiseaseDataset/processedCleveland.csv',header=None)
df = df.set_axis(feature_names, axis=1)

# missing values are marked as ?
missing_values_df = df[df.eq('?').any(axis=1)]
print('Missing attributes:')
display(missing_values_df)

# replacing missing values
# thal values are replaces with 3, normal
df['thal'] = df['thal'].replace('?',3)
# ca values are replaces with 0, no heart vessels are marked

df['ca'] = df['ca'].replace('?',0)

# missing values are marked as ?
missing_values_df = df[df.eq('?').any(axis=1)]
print('Missing attributes 2:')
display(missing_values_df)

print('Cleveland data frame:')
display(df)
display(df.describe())
display(df.info())
```

```python
df.to_csv('./HeartDiseaseDataset/ImportedDataset.csv')
```
